```python
from langchain_community.document_loaders import TextLoader
loader=TextLoader(r'C:\Users\admin\Desktop\GEN AI\TEXT.txt')
loader
```

```python
text=loader.load()
text
```

```python
print(text[0])
```

*SPLITTER TYPE ONE*

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
txts=RecursiveCharacterTextSplitter(chunk_size=100,chunk_overlap=10)
final=txts.split_documents(text)
final
```

```python
print(final[0])
print(final[1])
```

*SPLITTER TYPE TWO*

```python
from langchain_text_splitters import CharacterTextSplitter
#create a text splitter
txtsplit=CharacterTextSplitter(separator="\n",chunk_size=100,chunk_overlap=20)

#split the documents
tn= txtsplit.split_documents(text)
```
