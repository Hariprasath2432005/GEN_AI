```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
```

```python
from langchain_groq import ChatGroq
from langchain_community.vectorstores import FAISS
```

```python
import os 
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ["HUGGING_FACE"]=os.getenv("HUGGING_FACE")
```

```python
loader=TextLoader(r'C:\Users\admin\Desktop\GEN AI\TEXT.txt')
docs= loader.load()
print(docs)
```

```python
text_splitter=CharacterTextSplitter(chunk_size=1000,chunk_overlap=20)
documents=text_splitter.split_documents(docs)
print(documents)
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

# Use an embedding model instead of a chat model
emb=HuggingFaceEmbeddings(model_name="all-miniLM-L6-v2")
```

```python
# Now create the vector store

db=FAISS.from_documents(documents,emb)
db.save_local("faiss")
```

```python
query="what is Sentence Segmentation?"
docs=db.similarity_search(query)
docs
```

```python
docs[0].page_content
```

```python
docs[1].page_content
```

```python
#retriver
retriver=db.as_retriever()
retriver.invoke(query)
```

```python
#similarity score (manhatten_score)
doc_and_score=db.similarity_search_with_score("what is text preprocessing?")
print(doc_and_score)
```

```python
emb_vector=emb.embed_query("what is text preprocessing?")
emb_vector
```

```python
doc_score=db.similarity_search_by_vector(emb_vector)
print(doc_score)
```
